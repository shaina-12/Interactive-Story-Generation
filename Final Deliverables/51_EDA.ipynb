```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import nltk
nltk.download('punkt')
```

# Adventure Stories

```python
a_train = pd.read_csv('/content/Adventure Stories Train.csv')
a_val = pd.read_csv('/content/Adventure Stories Val.csv')
a_test = pd.read_csv('/content/Adventure Stories Test.csv')
```

```python
a_train.head()
```

```python
a_val.head()
```

```python
a_test.head()
```

```python
# taken reference from https://github.com/shaina-12/CSE508_Winter2024_A2_MT23139/blob/main/CSE508_Winter2024_A2_MT23139_Code_File.ipynb
# taken reference from https://github.com/shaina-12/CSE508_Winter2024_A1_MT23139/blob/main/CSE508_Winter2024_A1_MT23139_Code_File.py

def tokenize_words(text):
    return nltk.word_tokenize(text, language="english")

# taken reference from https://github.com/shaina-12/CSE508_Winter2024_A1_MT23139/blob/main/CSE508_Winter2024_A1_MT23139_Code_File.py
# taken reference from https://www.geeksforgeeks.org/python-remove-punctuation-from-string/

def remove_punctuation(text):
    p_text = ""
    punc = '''!?@#~-[]{}+=//\$()%;:'"\,<>./^&*_'''
    for i in range(len(text)):
        if text[i] not in punc:
            p_text += text[i]
    return p_text

def processing(text):
    #print(text)
    text = remove_punctuation(text).lower()
    text = tokenize_words(text)
    return text
```

```python
a_train_story = list(a_train['Story'])
a_val_story = list(a_val['Story'])
a_test_story = list(a_test['Story'])
```

```python
a_train_tokens_length = []
a_val_tokens_length = []
a_test_tokens_length = []

for i in range(len(a_train_story)):
  a_train_tokens_length.append(len(processing(a_train_story[i])))

for i in range(len(a_val_story)):
  a_val_tokens_length.append(len(processing(a_val_story[i])))

for i in range(len(a_test_story)):
  a_test_tokens_length.append(len(processing(a_test_story[i])))
```

```python
tok_length = np.array([max(a_train_tokens_length), max(a_val_tokens_length), max(a_test_tokens_length)])
x = np.arange(3)
plt.bar(x,tok_length)
labels = ['train set', 'validation set', 'test set']
plt.xticks(x, labels)
plt.xlabel('Data Spit')
plt.ylabel('Max. No. of Tokens in Story')
plt.title('Max. No. of Tokens in Adventure Stories of Each Split')
plt.show()
```

```python
print(tok_length)
```
